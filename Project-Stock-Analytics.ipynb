```python
import yfinance as yf

# Extracting Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_stock_data = tesla.history(period="max")
```

```python
tesla_stock_data.head()
```

```python
tesla_stock_data.tail()
```

```python
# Extracting GameStop stock data
gme = yf.Ticker("GME")
gme_stock_data = gme.history(period="max")

```

```python
# Displaying the first five rows
gme_stock_data.head()
```

```python
# Displaying the last five rows
gme_stock_data.tail()
```

```python
import matplotlib.pyplot as plt

# Function to make graph
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'])
    plt.xlabel('Date')
    plt.ylabel('Closing Price ($)')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Graphing Tesla stock data
title = "Tesla Stock Data"
make_graph(tesla_stock_data, title)
```
